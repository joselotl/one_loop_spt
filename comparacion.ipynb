```python
def name_reb(name_jos):
    name_jos = name_jos.split('_')
    fnl = name_jos[1]
    if fnl == '0':
        fnl_reb = '0'
    elif fnl == '-6':
        fnl_reb = '6'
    elif fnl == '4.2':
        fnl_reb = '4'
    name_jos = name_jos[3].split('/')
    gnl = name_jos[0]
    if gnl == '-12.3e4':
        gnl_reb = '12'
    elif gnl == '0':
        gnl_reb = '0'
    elif gnl == '0.7e4':
        gnl_reb = '7'
    base_name = name_jos[1].split('.')[0]
    if base_name[:3] == 'pkn':
        return 'newt'
    name_reb = '/home/josue/Dropbox/ PhD work/Paper Juan Carlos, Josue, Rebeca/Integrales-newlimits/'
    name_reb += base_name + 'fnl' + fnl_reb + 'gnl' + gnl_reb + '.txt'
    return name_reb
        
```
